```python
import numpy as np
import matplotlib.pyplot as plt
import pandas as pd
from analysis_functions import Analysis
```

```python
acrobot_baseline_no_in_out = "../../results/acrobot_replication/acrobot_skolik_baseline_no_in_out"
acrobot_baseline_no_in = "../../results/acrobot_replication/acrobot_skolik_baseline_no_in"
acrobot_baseline_no_out = "../../results/acrobot_replication/acrobot_skolik_baseline_no_out"
acrobot_baseline = "../../results/acrobot_replication/acrobot_skolik_baseline"
acrobot_datareup = "../../results/thesis/acrobot/acrobot_skolik_5layers"
acrobot_datareup_no_in_out = "../../results/acrobot_replication/acrobot_skolik_datareup_no_in_out"
acrobot_datareup_no_out = "../../results/acrobot_replication/acrobot_skolik_datareup_no_out"
acrobot_datareup_no_in = "../../results/acrobot_replication/acrobot_skolik_datareup_no_in"
```

```python
acrobot_baseline = Analysis(acrobot_baseline)
acrobot_baseline_no_in = Analysis(acrobot_baseline_no_in)
acrobot_baseline_no_out = Analysis(acrobot_baseline_no_out)
acrobot_baseline_no_in_out = Analysis(acrobot_baseline_no_in_out)
acrobot_datareup = Analysis(acrobot_datareup)
acrobot_datareup_no_in_out = Analysis(acrobot_datareup_no_in_out)
acrobot_datareup_no_out = Analysis(acrobot_datareup_no_out)
acrobot_datareup_no_in = Analysis(acrobot_datareup_no_in)
```

```python
acrobot_baseline_return = acrobot_baseline.get_rewards()
acrobot_baseline_no_in_return = acrobot_baseline_no_in.get_rewards()
acrobot_baseline_no_out_return = acrobot_baseline_no_out.get_rewards()
acrobot_baseline_no_in_out_return = acrobot_baseline_no_in_out.get_rewards()
acrobot_datareup_return = acrobot_datareup.get_rewards()
acrobot_datareup_no_in_out_return = acrobot_datareup_no_in_out.get_rewards()
acrobot_datareup_no_out_return = acrobot_datareup_no_out.get_rewards()
acrobot_datareup_no_in_return = acrobot_datareup_no_in.get_rewards()

acrobot_baseline_mean = np.mean(acrobot_baseline_return, axis=0)
acrobot_baseline_no_in_mean = np.mean(acrobot_baseline_no_in_return, axis=0)
acrobot_baseline_no_out_mean = np.mean(acrobot_baseline_no_out_return, axis=0)
acrobot_baseline_no_in_out_mean = np.mean(acrobot_baseline_no_in_out_return, axis=0)
acrobot_datareup_mean = np.mean(acrobot_datareup_return, axis=0)
acrobot_datareup_no_in_out_mean = np.mean(acrobot_datareup_no_in_out_return, axis=0)
acrobot_datareup_no_out_mean = np.mean(acrobot_datareup_no_out_return, axis=0)
acrobot_datareup_no_in_mean = np.mean(acrobot_datareup_no_in_return, axis=0)

acrobot_baseline_std = np.std(acrobot_baseline_return, axis=0)
acrobot_baseline_no_in_std = np.std(acrobot_baseline_no_in_return, axis=0)
acrobot_baseline_no_out_std = np.std(acrobot_baseline_no_out_return, axis=0)
acrobot_baseline_no_in_out_std = np.std(acrobot_baseline_no_in_out_return, axis=0)
acrobot_datareup_std = np.std(acrobot_datareup_return, axis=0)
acrobot_datareup_no_in_out_std = np.std(acrobot_datareup_no_in_out_return, axis=0)
acrobot_datareup_no_out_std = np.std(acrobot_datareup_no_out_return, axis=0)
acrobot_datareup_no_in_std = np.std(acrobot_datareup_no_in_return, axis=0)
```

```python
plt.plot(acrobot_baseline_mean, label="baseline")
plt.fill_between(range(len(acrobot_baseline_mean)), acrobot_baseline_mean - acrobot_baseline_std,
                    acrobot_baseline_mean + acrobot_baseline_std, alpha=0.2)
plt.plot(acrobot_baseline_no_in_mean, label="no input scaling")
plt.fill_between(range(len(acrobot_baseline_no_in_mean)), acrobot_baseline_no_in_mean - acrobot_baseline_no_in_std,
                    acrobot_baseline_no_in_mean + acrobot_baseline_no_in_std, alpha=0.2)
plt.plot(acrobot_baseline_no_out_mean, label="no output scaling")
plt.fill_between(range(len(acrobot_baseline_no_out_mean)), acrobot_baseline_no_out_mean - acrobot_baseline_no_out_std,
                    acrobot_baseline_no_out_mean + acrobot_baseline_no_out_std, alpha=0.2)
plt.plot(acrobot_baseline_no_in_out_mean, label="no input or output scaling")
plt.fill_between(range(len(acrobot_baseline_no_in_out_mean)), acrobot_baseline_no_in_out_mean - acrobot_baseline_no_in_out_std,
                    acrobot_baseline_no_in_out_mean + acrobot_baseline_no_in_out_std, alpha=0.2)
plt.legend()
plt.xlabel("Episode")
plt.ylabel("Return")
plt.title("Acrobot")
plt.show()

plt.plot(acrobot_datareup_mean, label="data reup")
plt.fill_between(range(len(acrobot_datareup_mean)), acrobot_datareup_mean - acrobot_datareup_std,
                    acrobot_datareup_mean + acrobot_datareup_std, alpha=0.2)
plt.plot(acrobot_datareup_no_in_mean, label="data reup no input scaling")
plt.fill_between(range(len(acrobot_datareup_no_in_mean)), acrobot_datareup_no_in_mean - acrobot_datareup_no_in_std,
                    acrobot_datareup_no_in_mean + acrobot_datareup_no_in_std, alpha=0.2)
plt.plot(acrobot_datareup_no_out_mean, label="data reup no output scaling")
plt.fill_between(range(len(acrobot_datareup_no_out_mean)), acrobot_datareup_no_out_mean - acrobot_datareup_no_out_std,
                    acrobot_datareup_no_out_mean + acrobot_datareup_no_out_std, alpha=0.2)
plt.plot(acrobot_datareup_no_in_out_mean, label="data reup no input or output scaling")
plt.fill_between(range(len(acrobot_datareup_no_in_out_mean)), acrobot_datareup_no_in_out_mean - acrobot_datareup_no_in_out_std,
                    acrobot_datareup_no_in_out_mean + acrobot_datareup_no_in_out_std, alpha=0.2)
plt.legend()
plt.xlabel("Episode")
plt.ylabel("Return")
plt.title("Acrobot")
plt.show()
```
