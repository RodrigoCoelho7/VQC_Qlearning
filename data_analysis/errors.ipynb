Now let's see if the baseline model is capable of learning when input scaling is used.

```python
baseline_input_scaling_path = "../../results/testing/baseline_input_scaling"

baseline_input_scaling = Analysis(baseline_input_scaling_path)
```

```python
baseline_input_scaling_return = baseline_input_scaling.get_moving_average(25)
```

```python
baseline_input_scaling_mean = np.mean(baseline_input_scaling_return, axis=0)
baseline_input_scaling_std = np.std(baseline_input_scaling_return, axis=0)

# plot the mean of the moving averages for each agent with different colors
fig, axs = plt.subplots(1, 1, tight_layout = True, figsize=(12, 5.5), sharey = True)
axs.plot(baseline_input_scaling_mean, color='orange', linewidth=1.5, label='Baseline Input Scaling')
axs.plot(means_baseline, color='blue', linewidth=1.5, label='Baseline')


# add shaded regions for the standard deviation of the moving averages
axs.fill_between(range(len(baseline_input_scaling_mean)), baseline_input_scaling_mean-baseline_input_scaling_std, baseline_input_scaling_mean+baseline_input_scaling_std, color='orange', alpha=0.15)
axs.fill_between(range(len(means_baseline)), means_baseline-stds_baseline, means_baseline+stds_baseline, color='blue', alpha=0.15)

# add labels and title
axs.set_xlabel('Episode', fontsize=14)
axs.set_ylabel('Return', fontsize=14)
axs.legend()

# show the plot
plt.show()
```

I found out why my baseline model wasnt working. I wasnt using input scaling but I should have used it. Because using input scaling even if the input encoding is done only once changes the frequency spectrum. It doesnt add frequencies to the spectrum but scales the existing frequencies so that they match the frequencies of the output function. This is why the model wasnt learning, probably the frequency of the output function wand the frequencies of the model were unmatched and all we needed was for input scaling to serve as a frequency matcher.
