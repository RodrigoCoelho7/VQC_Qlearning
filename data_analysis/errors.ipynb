```python
import matplotlib.pyplot as plt
import numpy as np
import pandas as pd
from analysis_functions import Analysis
```

```python
skolik_linear_path = "../../results/testing/skolik_linear"
skolik_tanh_path = "../../results/testing/skolik_tanh"

skolik_linear = Analysis(skolik_linear_path)
skolik_tanh = Analysis(skolik_tanh_path)
```

```python
skolik_linear_return = skolik_linear.get_moving_average(25)
skolik_tanh_return = skolik_tanh.get_moving_average(25)
```

```python
# calculate the mean and standard deviation of the moving averages for each agent
means_skolik_linear = np.mean(skolik_linear_return, axis=0)
stds_skolik_linear = np.std(skolik_linear_return, axis=0)
means_skolik_tanh = np.mean(skolik_tanh_return, axis=0)
stds_skolik_tanh = np.std(skolik_tanh_return, axis=0)



# plot the mean of the moving averages for each agent with different colors
fig, axs = plt.subplots(1, 1, tight_layout = True, figsize=(12, 5.5), sharey = True)
axs.plot(means_skolik_linear, color='orange', linewidth=1.5, label='Skolik Linear')
axs.plot(means_skolik_tanh, color='blue', linewidth=1.5, label='Skolik Tanh')


# add shaded regions for the standard deviation of the moving averages
axs.fill_between(range(len(means_skolik_linear)), means_skolik_linear-stds_skolik_linear, means_skolik_linear+stds_skolik_linear, color='orange', alpha=0.15)
axs.fill_between(range(len(means_skolik_tanh)), means_skolik_tanh-stds_skolik_tanh, means_skolik_tanh+stds_skolik_tanh, color='blue', alpha=0.15)

# add labels and title
axs.set_xlabel('Episode', fontsize=14)
axs.set_ylabel('Return', fontsize=14)
axs.legend()

# show the plot
plt.show()
```

If anything, the tanh activation seems to lead to a worse performance. So, the dropoff in performance didnt come from the activation.

I found out why my baseline model wasnt working. I wasnt using input scaling but I should have used it. Because using input scaling even if the input encoding is done only once changes the frequency spectrum. It doesnt add frequencies to the spectrum but scales the existing frequencies so that they match the frequencies of the output function. This is why the model wasnt learning, probably the frequency of the output function wand the frequencies of the model were unmatched and all we needed was for input scaling to serve as a frequency matcher.
