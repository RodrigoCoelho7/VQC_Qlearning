```python
import numpy as np
import pandas as pd
from analysis_functions import Analysis
import matplotlib.pyplot as plt
```

```python
uqc_4qubits_frozenlake_path = "../../results/frozenlake/uqc_4qubits"

uqc_4qubits_frozenlake = Analysis(uqc_4qubits_frozenlake_path)
```

```python
uqc_4qubits_frozenlake_return = uqc_4qubits_frozenlake.get_rewards()

uqc_4qubits_frozenlake_mean_return = np.mean(uqc_4qubits_frozenlake_return, axis=0)

uqc_4qubits_frozenlake_std_return = np.std(uqc_4qubits_frozenlake_return, axis=0)

uqc_4qubits_frozenlake_mean_return_moving_average = uqc_4qubits_frozenlake.get_moving_average(25)

uqc_4qubits_final = np.mean(uqc_4qubits_frozenlake_mean_return_moving_average, axis=0)
```

```python
plt.plot(uqc_4qubits_final, label="UQC 4 qubits")
plt.fill_between(np.arange(len(uqc_4qubits_frozenlake_mean_return)), uqc_4qubits_frozenlake_mean_return - uqc_4qubits_frozenlake_std_return, uqc_4qubits_frozenlake_mean_return + uqc_4qubits_frozenlake_std_return, alpha=0.2)
plt.legend()
plt.show()
```

```python
frozenlake_mean_gradients, frozenlake_std_gradients, frozenlake_max, frozenlake_max_index, frozenlake_min, frozenlake_min_index = uqc_4qubits_frozenlake.calculate_mean_variance_gradients(return_max = True, return_min = True)
```

```python
frozenlake_mean_gradients_smoothed = pd.Series(frozenlake_mean_gradients).rolling(250).mean()
```

```python
plt.plot(frozenlake_mean_gradients_smoothed, label="UQC 4 qubits")
```
