```python
import numpy as np
import matplotlib.pyplot as plt
import pandas as pd
from analysis_functions import Analysis
```

```python
model_path = "../../results/thesis_organized/1.1/skolik_datareup"

model = Analysis(model_path)
```

```python
gradients = model.get_gradients()

gradients[0][0]
```

```python
def flatten_gradients(gradients):
    for i in range(len(gradients)):
        for j in range(len(gradients[i])):
            gradients[i][j] = np.concatenate([lista.flatten() for lista in gradients[i][j]], axis = 0)

min_length = min([len(gradients[i]) for i in range(len(gradients))])

gradients = [gradients[i][:min_length] for i in range(len(gradients))]

flatten_gradients(gradients)

gradients_array = np.array(gradients)
print(gradients_array.shape)
```

```python
# The first 48 gradients belong to the rotational parameters. Let's start by taking the first one and seeing how it evolves over time.
rotational_gradients = gradients_array[:, :, :48]
print(rotational_gradients.shape)

# Let's take the first gradient and see how it evolves over time.
first_gradient = rotational_gradients[:, :, 0]
print(first_gradient.shape)

# Let's plot the first gradient over time.
plt.plot(first_gradient.T)
plt.show()

# Now let's calculate the variance of the first gradient over time
variance = np.var(first_gradient, axis = 0)
plt.plot(variance)
plt.show()

#Now let's do the same but for the last gradient
last_gradient = rotational_gradients[:, :, -22]
plt.plot(last_gradient.T)
plt.show()
variance = np.var(last_gradient, axis = 0)
plt.plot(variance)
plt.show()

rotational_gradients_variance = np.var(rotational_gradients, axis = 0)
print(rotational_gradients_variance.shape)

plt.plot(rotational_gradients_variance.T)
plt.show()
```
