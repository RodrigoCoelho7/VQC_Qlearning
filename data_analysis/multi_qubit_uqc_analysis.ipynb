# 1) Performance of the models

## 1.1) Performance of the Full Encoding Models

```python
moving_average_single_qubit = analysis_single_qubit.get_moving_average(25)
moving_average_2q2l_full = analysis_2q2l_full.get_moving_average(25)
moving_average_2q3l_full = analysis_2q3l_full.get_moving_average(25)
moving_average_2q4l_full = analysis_2q4l_full.get_moving_average(25)
moving_average_2q5l_full = analysis_2q5l_full.get_moving_average(25)
moving_average_4q2l_full = analysis_4q2l_full.get_moving_average(25)
moving_average_4q3l_full = analysis_4q3l_full.get_moving_average(25)
moving_average_4q4l_full = analysis_4q4l_full.get_moving_average(25)
moving_average_4q5l_full = analysis_4q5l_full.get_moving_average(25)
```

First, let's compare the performance of the same number of layers for 1 qubit, 2 qubits and 4 qubits

```python
# calculate the mean and standard deviation of the moving averages for each agent
means_single_qubit = np.mean(moving_average_single_qubit, axis=0)
stds_single_qubit = np.std(moving_average_single_qubit, axis=0)
means_2q5l_full = np.mean(moving_average_2q5l_full, axis=0)
stds_2q5l_full = np.std(moving_average_2q5l_full, axis=0)
means_4q_5l_full = np.mean(moving_average_4q5l_full, axis=0)
stds_4q_5l_full = np.std(moving_average_4q5l_full, axis=0)


# plot the mean of the moving averages for each agent with different colors
fig, axs = plt.subplots(1, 1, tight_layout = True, figsize=(12, 5.5), sharey = True)
axs.plot(means_single_qubit, color='blue', linewidth=1.5, label='Single-Qubit 5 layers')
axs.plot(means_2q5l_full, color='red', linewidth=1.5, label='Two-Qubits 5 layers')
axs.plot(means_4q_5l_full, color='green', linewidth=1.5, label='Four-Qubits 5 layers')

area_single_qubit = np.trapz(means_single_qubit[25:1000])
area_2q5l_full = np.trapz(means_2q5l_full[25:])
area_4q_5l_full = np.trapz(means_4q_5l_full[25:])


# add shaded regions for the standard deviation of the moving averages
axs.fill_between(range(len(means_single_qubit)), means_single_qubit-stds_single_qubit, means_single_qubit+stds_single_qubit, color='blue', alpha=0.15)
axs.fill_between(range(len(means_2q5l_full)), means_2q5l_full-stds_2q5l_full, means_2q5l_full+stds_2q5l_full, color='red', alpha=0.15)
axs.fill_between(range(len(means_4q_5l_full)), means_4q_5l_full-stds_4q_5l_full, means_4q_5l_full+stds_4q_5l_full, color='green', alpha=0.15)

# add labels and title
axs.set_xlabel('Episode', fontsize=14)
axs.set_ylabel('Return', fontsize=14)
axs.legend()

# show the plot
plt.show()
```

```python
print("Area under the curve for single qubit: ", area_single_qubit)
print("Area under the curve for 2 qubits: ", area_2q5l_full)
print("Area under the curve for 4 qubits: ", area_4q_5l_full)
print("\n")
print("Ratio between 2 qubits and single qubit: ", area_2q5l_full/area_single_qubit)
print("Ratio between 4 qubits and single qubit: ", area_4q_5l_full/area_single_qubit)
```

There doesn't seem to exist an advantage in increasing the number of qubits. Actually, if anything, the performance seems to decrease. This is very interesting because the number of parameters increases proportionally to the number of qubits, so in theory these models should be more expressive than the 1-qubit models.

Now let's see how the performance of the 2 qubits varies with the number of layers

```python
# calculate the mean and standard deviation of the moving averages for each agent
means_2q2l_full = np.mean(moving_average_2q2l_full, axis=0)
stds_2q2l_full = np.std(moving_average_2q2l_full, axis=0)
means_2q3l_full = np.mean(moving_average_2q3l_full, axis=0)
stds_2q3l_full = np.std(moving_average_2q3l_full, axis=0)
means_2q4l_full = np.mean(moving_average_2q4l_full, axis=0)
stds_2q4l_full = np.std(moving_average_2q4l_full, axis=0)
means_2q5l_full = np.mean(moving_average_2q5l_full, axis=0)
stds_2q5l_full = np.std(moving_average_2q5l_full, axis=0)


# plot the mean of the moving averages for each agent with different colors
fig, axs = plt.subplots(1, 1, tight_layout = True, figsize=(12, 5.5), sharey = True)
axs.plot(means_2q2l_full, color='blue', linewidth=1.5, label='2 qubits and 2 layers')
axs.plot(means_2q3l_full, color='red', linewidth=1.5, label='2 qubits and 3 layers')
axs.plot(means_2q4l_full, color='green', linewidth=1.5, label='2 qubits and 4 layers')
axs.plot(means_2q5l_full, color='orange', linewidth=1.5, label='2 qubits and 5 layers')


# add shaded regions for the standard deviation of the moving averages
#axs.fill_between(range(len(means_single_qubit)), means_single_qubit-stds_single_qubit, means_single_qubit+stds_single_qubit, color='blue', alpha=0.15)
#axs.fill_between(range(len(means_2q5l_full)), means_2q5l_full-stds_2q5l_full, means_2q5l_full+stds_2q5l_full, color='red', alpha=0.15)
#axs.fill_between(range(len(means_4q_5l_full)), means_4q_5l_full-stds_4q_5l_full, means_4q_5l_full+stds_4q_5l_full, color='green', alpha=0.15)

# add labels and title
axs.set_xlabel('Episode', fontsize=14)
axs.set_ylabel('Return', fontsize=14)
axs.legend()

# show the plot
plt.show()
```

It seems like these models saturate faster than the 1-qubit models. This is probably due to the fact that these models have double the amount of parameters than the 1-qubit models, and thus might be more expressive, consequently having reduced trainability (expressivity-trainability tradeoff).

Let's do the same study but for the 4 qubit models.

```python
# calculate the mean and standard deviation of the moving averages for each agent
means_4q2l_full = np.mean(moving_average_4q2l_full, axis=0)
stds_4q2l_full = np.std(moving_average_4q2l_full, axis=0)
means_4q3l_full = np.mean(moving_average_4q3l_full, axis=0)
stds_4q3l_full = np.std(moving_average_4q3l_full, axis=0)
means_4q4l_full = np.mean(moving_average_4q4l_full, axis=0)
stds_4q4l_full = np.std(moving_average_4q4l_full, axis=0)
means_4q5l_full = np.mean(moving_average_4q5l_full, axis=0)
stds_4q5l_full = np.std(moving_average_4q5l_full, axis=0)


# plot the mean of the moving averages for each agent with different colors
fig, axs = plt.subplots(1, 1, tight_layout = True, figsize=(12, 5.5), sharey = True)
axs.plot(means_4q2l_full, color='blue', linewidth=1.5, label='4 qubits and 2 layers')
axs.plot(means_4q3l_full, color='red', linewidth=1.5, label='4 qubits and 3 layers')
axs.plot(means_4q4l_full, color='green', linewidth=1.5, label='4 qubits and 4 layers')
axs.plot(means_4q5l_full, color='orange', linewidth=1.5, label='4 qubits and 5 layers')


# add shaded regions for the standard deviation of the moving averages
#axs.fill_between(range(len(means_single_qubit)), means_single_qubit-stds_single_qubit, means_single_qubit+stds_single_qubit, color='blue', alpha=0.15)
#axs.fill_between(range(len(means_4q5l_full)), means_4q5l_full-stds_4q5l_full, means_4q5l_full+stds_4q5l_full, color='red', alpha=0.15)
#axs.fill_between(range(len(means_4q_5l_full)), means_4q_5l_full-stds_4q_5l_full, means_4q_5l_full+stds_4q_5l_full, color='green', alpha=0.15)

# add labels and title
axs.set_xlabel('Episode', fontsize=14)
axs.set_ylabel('Return', fontsize=14)
axs.legend()

# show the plot
plt.show()
```

Interestingly enough, while the performance of the 4 qubit models is much better for the two layers case than the performance of the 2 qubit models, the performance for a higher number of layers seems to be worse. Why could this be? One possible explanation is once again the expressivity-trainability tradeoff. These models have 4 times the number of parameters of the 1-qubit models and twice that of the 2-qubit models. Hence, they might be harder to train? To further confirm these claims, we need to check the gradients of the cost function with respect to the parameters.

Before doing that, let's compare the performance of each one of the best models for each number of qubits.

- Single-qubit case: 5 layers
- Two-qubit case: 3 layers
- Four Qubit case: 5 layers

```python
means_single_qubit = np.mean(moving_average_single_qubit, axis=0)
stds_single_qubit = np.std(moving_average_single_qubit, axis=0)
means_2q3l_full = np.mean(moving_average_2q3l_full, axis=0)
stds_2q3l_full = np.std(moving_average_2q3l_full, axis=0)
means_4q5l_full = np.mean(moving_average_4q5l_full, axis=0)
stds_4q5l_full = np.std(moving_average_4q5l_full, axis=0)


fig, axs = plt.subplots(1, 1, tight_layout = True, figsize=(12, 5.5), sharey = True)
axs.plot(means_single_qubit, color='red', linewidth=1.5, label='Best 1-qubit model')
axs.plot(means_2q3l_full, color='blue', linewidth=1.5, label='Best 2-qubit model')
axs.plot(means_4q5l_full, color='green', linewidth=1.5, label='Best 4-qubit model')

axs.set_xlabel('Episode', fontsize=14)
axs.set_ylabel('Return', fontsize=14)
axs.legend()

# show the plot
plt.show()
```

Interestingly enough, the best 2-qubit model seems to perform the best when compared to the other models with different numbers of qubits. Could it possibly be due to the fact that the 2-qubit model has a number of parameters and qubits that are in a "golden zone"? Maybe there is certain range in terms of parameters where, once past this range, the model could become harder to train and this would in consequence affect the performance of the model. I am not sure, but this is a very interesting result.

Once again, using the MSE, we have huge gradients. This could be explained by a couple of factors, related to the fact that it is normal for the loss in DQN to actually increase as the agent learns and improves:

- First, as the agent leanrs, the Q-values become harder and harder to estimate, since the episodes become longer. It is a lot harder to estimate the value of an action that could yield rewards for the next 200 steps than it is for an action that will yield shorter-term rewards.

- Secondly, as the agent learns, the variance in the Q-values also increases.

- Lastly, eventually the replay buffer becomes full with similar (well-performing) experiences, which means that it will forget different experiences, but those are also very important for the training of a RL agent. We can't see that, because I stop training once the environment is solved. However, If I kept them training, we would probably see the gradients getting larger and larger and the agent unlearning.
