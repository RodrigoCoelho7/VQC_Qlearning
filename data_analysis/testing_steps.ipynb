```python
skolik_rewards = skolik.get_rewards()
skolik_steps_per_episode = skolik.get_steps_per_episode()
```

```python
cumulative_steps = [[],[],[],[],[]]

for i in range(len(skolik_steps_per_episode)):
    cumulative_steps[i].append(skolik_steps_per_episode[i][0])
    for j in range(1,len(skolik_steps_per_episode[i])):
        cumulative_steps[i].append(cumulative_steps[i][-1] + skolik_steps_per_episode[i][j])
```

```python
plt.plot(cumulative_steps[0], skolik_rewards[0], label="Skolik 0")
plt.plot(cumulative_steps[1], skolik_rewards[1], label="Skolik 1")
plt.plot(cumulative_steps[2], skolik_rewards[2], label="Skolik 2")
plt.plot(cumulative_steps[3], skolik_rewards[3], label="Skolik 3")
plt.plot(cumulative_steps[4], skolik_rewards[4], label="Skolik 4")
```
