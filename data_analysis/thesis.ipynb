```python
import random
import numpy as np
import matplotlib.pyplot as plt
import pandas as pd
from analysis_functions import Analysis
```

```python
skolik_datareup_path = "../../results/thesis/1.1/skolik_datareup"
skolik_datareup = Analysis(skolik_datareup_path)
gradients = skolik_datareup.get_gradients()

selected_indices = random.sample(range(len(gradients)), 3)
selected_gradients = [gradients[i] for i in selected_indices]

#5,7,6
```

```python
def calculate_norm_of_lists_of_gradients(lists_of_gradients):
    for i in range(len(lists_of_gradients)):
        for j in range(len(lists_of_gradients[i])):
            lists_of_gradients[i][j] = np.linalg.norm(np.concatenate([lista.flatten() for lista in lists_of_gradients[i][j]], axis = 0))
        
```

```python
calculate_norm_of_lists_of_gradients(selected_gradients)
```

```python
for i in range(len(selected_gradients)):
    selected_gradients[i] = pd.Series(selected_gradients[i]).rolling(250).mean()
```

```python
skolik_datareup_returns = skolik_datareup.get_rewards()
skolik_datareup_steps = skolik_datareup.get_steps_per_episode()

skolik_datareup_selected_returns = [skolik_datareup_returns[i] for i in selected_indices]
skolik_datareup_selected_steps = [skolik_datareup_steps[i] for i in selected_indices]
```

```python
for i in range(len(skolik_datareup_selected_returns)):
    skolik_datareup_selected_returns[i] = pd.Series(skolik_datareup_selected_returns[i]).rolling(10).mean()
```

```python
def calculate_cumulative_steps(list_of_steps_per_episode):
    # Step 1: Initialize the cumulative_steps list
    cumulative_steps = [list_of_steps_per_episode[0]]

    # Step 2: Calculate cumulative steps for each episode
    for i in range(1, len(list_of_steps_per_episode)):
        cumulative_steps.append(cumulative_steps[-1] + list_of_steps_per_episode[i])
    return cumulative_steps
```

```python
cumulative_steps = [calculate_cumulative_steps(skolik_datareup_selected_steps[0]), calculate_cumulative_steps(skolik_datareup_selected_steps[1]), calculate_cumulative_steps(skolik_datareup_selected_steps[2])]
```
