## 5.1) Replication of Skolik et.al's results

````python
skolik_baseline_path = "../../results/thesis/1.1/skolik_baseline"
skolik_datareup_path = "../../results/thesis/1.1/skolik_datareup"
skolik_baseline_no_in_path = "../../results/thesis/1.1/skolik_baseline_no_in"
skolik_datareup_no_in_path = "../../results/thesis/1.1/skolik_datareup_no_in"
skolik_baseline_no_out_path = "../../results/thesis/1.1/skolik_baseline_no_out"
skolik_datareup_no_out_path = "../../results/thesis/1.1/skolik_datareup_no_out"
skolik_baseline_no_in_out_path = "../../results/thesis/1.1/skolik_baseline_no_in_out"
skolik_datareup_no_in_out_path = "../../results/thesis/1.1/skolik_datareup_no_in_out"

skolik_baseline = Analysis(skolik_baseline_path)
skolik_datareup = Analysis(skolik_datareup_path)
skolik_baseline_no_in = Analysis(skolik_baseline_no_in_path)
skolik_datareup_no_in = Analysis(skolik_datareup_no_in_path)
skolik_baseline_no_out = Analysis(skolik_baseline_no_out_path)
skolik_datareup_no_out = Analysis(skolik_datareup_no_out_path)
skolik_baseline_no_in_out = Analysis(skolik_baseline_no_in_out_path)
skolik_datareup_no_in_out = Analysis(skolik_datareup_no_in_out_path)
````

````python
skolik_baseline_return = skolik_baseline.get_rewards()
skolik_datareup_return = skolik_datareup.get_rewards()
skolik_baseline_no_in_return = skolik_baseline_no_in.get_rewards()
skolik_datareup_no_in_return = skolik_datareup_no_in.get_rewards()
skolik_baseline_no_out_return = skolik_baseline_no_out.get_rewards()
skolik_datareup_no_out_return = skolik_datareup_no_out.get_rewards()
skolik_baseline_no_in_out_return = skolik_baseline_no_in_out.get_rewards()
skolik_datareup_no_in_out_return = skolik_datareup_no_in_out.get_rewards()

skolik_baseline_return_mean = np.mean(skolik_baseline_return, axis=0)
skolik_datareup_mean_rewards = np.mean(skolik_datareup_return, axis=0)
skolik_baseline_no_in_return_mean = np.mean(skolik_baseline_no_in_return, axis=0)
skolik_datareup_no_in_return_mean = np.mean(skolik_datareup_no_in_return, axis=0)
skolik_baseline_no_out_return_mean = np.mean(skolik_baseline_no_out_return, axis=0)
skolik_datareup_no_out_return_mean = np.mean(skolik_datareup_no_out_return, axis=0)
skolik_baseline_no_in_out_return_mean = np.mean(skolik_baseline_no_in_out_return, axis=0)
skolik_datareup_no_in_out_return_mean = np.mean(skolik_datareup_no_in_out_return, axis=0)

skolik_baseline_return_std = np.std(skolik_baseline_return, axis=0)
skolik_datareup_std_rewards= np.std(skolik_datareup_return, axis=0)
skolik_baseline_no_in_return_std = np.std(skolik_baseline_no_in_return, axis=0)
skolik_datareup_no_in_return_std = np.std(skolik_datareup_no_in_return, axis=0)
skolik_baseline_no_out_return_std = np.std(skolik_baseline_no_out_return, axis=0)
skolik_datareup_no_out_return_std = np.std(skolik_datareup_no_out_return, axis=0)
skolik_baseline_no_in_out_return_std = np.std(skolik_baseline_no_in_out_return, axis=0)
skolik_datareup_no_in_out_return_std = np.std(skolik_datareup_no_in_out_return, axis=0)
````

````python
fig,axs = plt.subplots(1,1,figsize=(7.5,5), tight_layout=True)

axs.plot(skolik_baseline_return_mean,color = "red", label="Baseline")
axs.plot(skolik_baseline_no_in_return_mean,color = "green", label="Baseline w/o input scaling")
axs.plot(skolik_baseline_no_out_return_mean,color = "orange", label="Baseline w/o output scaling")
axs.plot(skolik_baseline_no_in_out_return_mean,color = "purple", label="Baseline w/o input and output scaling")
axs.fill_between(np.arange(len(skolik_baseline_return_mean)), np.clip(skolik_baseline_return_mean - skolik_baseline_return_std,a_min = 0, a_max=200), np.clip(skolik_baseline_return_mean + skolik_baseline_return_std, a_min = None, a_max = 200), alpha=0.15, color="red")
axs.fill_between(np.arange(len(skolik_baseline_no_in_return_mean)), np.clip(skolik_baseline_no_in_return_mean - skolik_baseline_no_in_return_std,a_min=0,a_max=200), np.clip(skolik_baseline_no_in_return_mean + skolik_baseline_no_in_return_std, a_min = None, a_max = 200), alpha=0.15, color="green")
axs.fill_between(np.arange(len(skolik_baseline_no_out_return_mean)), np.clip(skolik_baseline_no_out_return_mean - skolik_baseline_no_out_return_std,a_min=0,a_max=200), np.clip(skolik_baseline_no_out_return_mean + skolik_baseline_no_out_return_std, a_min = None, a_max = 200), alpha=0.15, color="orange")
axs.fill_between(np.arange(len(skolik_baseline_no_in_out_return_mean)), np.clip(skolik_baseline_no_in_out_return_mean - skolik_baseline_no_in_out_return_std,a_min=0,a_max=200), np.clip(skolik_baseline_no_in_out_return_mean + skolik_baseline_no_in_out_return_std, a_min = None, a_max = 200), alpha=0.15, color="purple")
axs.set_xlabel("Episode")
axs.set_ylabel("Mean Return")
axs.legend(loc = "lower right")
plt.grid(True)
plt.savefig('../../graphs/skolik_baseline.pdf')  
plt.show()

fig,axs = plt.subplots(1,1,figsize=(7.5,5), tight_layout=True)

axs.plot(skolik_datareup_mean_rewards,color = "blue", label="Data Reuploading")
axs.plot(skolik_datareup_no_in_return_mean,color = "black", label="Data Reuploading w/o input scaling")
axs.plot(skolik_datareup_no_out_return_mean,color = "cyan", label="Data Reuploading w/o output scaling")
axs.plot(skolik_datareup_no_in_out_return_mean,color = "brown", label="Data Reuploading w/o input and output scaling")
axs.fill_between(np.arange(len(skolik_datareup_mean_rewards)), np.clip(skolik_datareup_mean_rewards - skolik_datareup_std_rewards,a_min=0,a_max=200), np.clip(skolik_datareup_mean_rewards + skolik_datareup_std_rewards, a_min = None, a_max = 200), alpha=0.15, color="blue")
axs.fill_between(np.arange(len(skolik_datareup_no_in_return_mean)), np.clip(skolik_datareup_no_in_return_mean - skolik_datareup_no_in_return_std,a_min=0,a_max=200), np.clip(skolik_datareup_no_in_return_mean + skolik_datareup_no_in_return_std, a_min = None, a_max = 200), alpha=0.15, color="black")
axs.fill_between(np.arange(len(skolik_datareup_no_out_return_mean)), np.clip(skolik_datareup_no_out_return_mean - skolik_datareup_no_out_return_std,a_min=0,a_max=200), np.clip(skolik_datareup_no_out_return_mean + skolik_datareup_no_out_return_std, a_min = None, a_max = 200), alpha=0.15, color="cyan")
axs.fill_between(np.arange(len(skolik_datareup_no_in_out_return_mean)), np.clip(skolik_datareup_no_in_out_return_mean - skolik_datareup_no_in_out_return_std,a_min=0,a_max=200), np.clip(skolik_datareup_no_in_out_return_mean + skolik_datareup_no_in_out_return_std, a_min = None, a_max = 200), alpha=0.15, color="brown")
axs.set_xlabel("Episode")
axs.set_ylabel("Mean Return")
axs.legend()
plt.grid(True)
plt.savefig('../../graphs/skolik_datareup.pdf')  
plt.show()

````

## 5.2.1) The Single-Qubit Universal Quantum Classifier

````python
one_qubit_uqc_path = "../../results/thesis/4.7/one_qubit_uqc"
two_qubits_full_path = "../../results/thesis/4.7/two_qubits_uqc_full"
two_qubits_partial_path = "../../results/thesis/4.7/two_qubits_uqc_partial"
four_qubits_full_path = "../../results/thesis/4.7/four_qubits_uqc_full"
four_qubits_partial_path = "../../results/thesis/4.7/four_qubits_uqc_partial"
two_qubits_uqc_full_product_state_path = "../../results/thesis/4.7_product_state/two_qubits_uqc_full"
two_qubits_uqc_partial_product_state_path = "../../results/thesis/4.7_product_state/two_qubits_uqc_partial"
four_qubits_uqc_full_product_state_path = "../../results/thesis/4.7_product_state/four_qubits_uqc_full"
four_qubits_uqc_partial_product_state_path = "../../results/thesis/4.7_product_state/four_qubits_uqc_partial"
skolik_datareup_path = "../../results/thesis/1.1/skolik_datareup"

one_qubit_uqc_analysis = Analysis(one_qubit_uqc_path)
two_qubits_full_analysis = Analysis(two_qubits_full_path)
two_qubits_partial_analysis = Analysis(two_qubits_partial_path)
four_qubits_full_analysis = Analysis(four_qubits_full_path)
four_qubits_partial_analysis = Analysis(four_qubits_partial_path)
two_qubits_uqc_full_product_state_analysis = Analysis(two_qubits_uqc_full_product_state_path)
two_qubits_uqc_partial_product_state_analysis = Analysis(two_qubits_uqc_partial_product_state_path)
four_qubits_uqc_full_product_state_analysis = Analysis(four_qubits_uqc_full_product_state_path)
four_qubits_uqc_partial_product_state_analysis = Analysis(four_qubits_uqc_partial_product_state_path)
skolik_datareup_analysis = Analysis(skolik_datareup_path)
````

````python
one_qubit_uqc_rewards = one_qubit_uqc_analysis.get_rewards()
two_qubits_full_rewards = two_qubits_full_analysis.get_rewards()
two_qubits_partial_rewards = two_qubits_partial_analysis.get_rewards()
four_qubits_full_rewards = four_qubits_full_analysis.get_rewards()
four_qubits_partial_rewards = four_qubits_partial_analysis.get_rewards()
two_qubits_uqc_full_product_state_rewards = two_qubits_uqc_full_product_state_analysis.get_rewards()
two_qubits_uqc_partial_product_state_rewards = two_qubits_uqc_partial_product_state_analysis.get_rewards()
four_qubits_uqc_full_product_state_rewards = four_qubits_uqc_full_product_state_analysis.get_rewards()
four_qubits_uqc_partial_product_state_rewards = four_qubits_uqc_partial_product_state_analysis.get_rewards()
skolik_datareup_rewards = skolik_datareup_analysis.get_rewards()

one_qubit_uqc_mean_rewards = np.mean(one_qubit_uqc_rewards, axis=0)
two_qubits_full_mean_rewards = np.mean(two_qubits_full_rewards, axis=0)
two_qubits_partial_mean_rewards = np.mean(two_qubits_partial_rewards, axis=0)
four_qubits_full_mean_rewards = np.mean(four_qubits_full_rewards, axis=0)
four_qubits_partial_mean_rewards = np.mean(four_qubits_partial_rewards, axis=0)
two_qubits_uqc_full_product_state_mean_rewards = np.mean(two_qubits_uqc_full_product_state_rewards, axis=0)
two_qubits_uqc_partial_product_state_mean_rewards = np.mean(two_qubits_uqc_partial_product_state_rewards, axis=0)
four_qubits_uqc_full_product_state_mean_rewards = np.mean(four_qubits_uqc_full_product_state_rewards, axis=0)
four_qubits_uqc_partial_product_state_mean_rewards = np.mean(four_qubits_uqc_partial_product_state_rewards, axis=0)
skolik_datareup_mean_rewards = np.mean(skolik_datareup_rewards, axis=0)


one_qubit_uqc_std_rewards = np.std(one_qubit_uqc_rewards, axis=0)
two_qubits_full_std_rewards = np.std(two_qubits_full_rewards, axis=0)
two_qubits_partial_std_rewards = np.std(two_qubits_partial_rewards, axis=0)
four_qubits_full_std_rewards = np.std(four_qubits_full_rewards, axis=0)
four_qubits_partial_std_rewards = np.std(four_qubits_partial_rewards, axis=0)
two_qubits_uqc_full_product_state_std_rewards = np.std(two_qubits_uqc_full_product_state_rewards, axis=0)
two_qubits_uqc_partial_product_state_std_rewards = np.std(two_qubits_uqc_partial_product_state_rewards, axis=0)
four_qubits_uqc_full_product_state_std_rewards = np.std(four_qubits_uqc_full_product_state_rewards, axis=0)
four_qubits_uqc_partial_product_state_std_rewards = np.std(four_qubits_uqc_partial_product_state_rewards, axis=0)
skolik_datareup_std_rewards = np.std(skolik_datareup_rewards, axis=0)
````

````python
one_qubit_uqc_norm, one_qubit_uqc_variance = one_qubit_uqc_analysis.calculate_mean_variance_gradients()
two_qubits_full_norm, two_qubits_full_variance = two_qubits_full_analysis.calculate_mean_variance_gradients()
two_qubits_partial_norm, two_qubits_partial_variance = two_qubits_partial_analysis.calculate_mean_variance_gradients()
four_qubits_full_norm, four_qubits_full_variance = four_qubits_full_analysis.calculate_mean_variance_gradients()
four_qubits_partial_norm, four_qubits_partial_variance = four_qubits_partial_analysis.calculate_mean_variance_gradients()
two_qubits_uqc_full_product_state_norm, two_qubits_uqc_full_product_state_variance = two_qubits_uqc_full_product_state_analysis.calculate_mean_variance_gradients()
two_qubits_uqc_partial_product_state_norm, two_qubits_uqc_partial_product_state_variance = two_qubits_uqc_partial_product_state_analysis.calculate_mean_variance_gradients()
four_qubits_uqc_full_product_state_norm, four_qubits_uqc_full_product_state_variance = four_qubits_uqc_full_product_state_analysis.calculate_mean_variance_gradients()
four_qubits_uqc_partial_product_state_norm, four_qubits_uqc_partial_product_state_variance = four_qubits_uqc_partial_product_state_analysis.calculate_mean_variance_gradients()
skolik_datareup_norm, skolik_datareup_variance = skolik_datareup_analysis.calculate_mean_variance_gradients()
````

````python
one_qubit_uqc_norm_smoothed = pd.Series(one_qubit_uqc_norm).rolling(100).mean()
two_qubits_full_norm_smoothed = pd.Series(two_qubits_full_norm).rolling(100).mean()
two_qubits_partial_norm_smoothed = pd.Series(two_qubits_partial_norm).rolling(100).mean()
four_qubits_full_norm_smoothed = pd.Series(four_qubits_full_norm).rolling(100).mean()
four_qubits_partial_norm_smoothed = pd.Series(four_qubits_partial_norm).rolling(100).mean()
two_qubits_uqc_full_product_state_norm_smoothed = pd.Series(two_qubits_uqc_full_product_state_norm).rolling(100).mean()
two_qubits_uqc_partial_product_state_norm_smoothed = pd.Series(two_qubits_uqc_partial_product_state_norm).rolling(100).mean()
four_qubits_uqc_full_product_state_norm_smoothed = pd.Series(four_qubits_uqc_full_product_state_norm).rolling(100).mean()
four_qubits_uqc_partial_product_state_norm_smoothed = pd.Series(four_qubits_uqc_partial_product_state_norm).rolling(100).mean()
skolik_datareup_norm_smoothed = pd.Series(skolik_datareup_norm).rolling(100).mean()

one_qubit_uqc_variance_smoothed = pd.Series(one_qubit_uqc_variance).rolling(100).mean()
two_qubits_full_variance_smoothed = pd.Series(two_qubits_full_variance).rolling(100).mean()
two_qubits_partial_variance_smoothed = pd.Series(two_qubits_partial_variance).rolling(100).mean()
four_qubits_full_variance_smoothed = pd.Series(four_qubits_full_variance).rolling(100).mean()
four_qubits_partial_variance_smoothed = pd.Series(four_qubits_partial_variance).rolling(100).mean()
two_qubits_uqc_full_product_state_variance_smoothed = pd.Series(two_qubits_uqc_full_product_state_variance).rolling(100).mean()
two_qubits_uqc_partial_product_state_variance_smoothed = pd.Series(two_qubits_uqc_partial_product_state_variance).rolling(100).mean()
four_qubits_uqc_full_product_state_variance_smoothed = pd.Series(four_qubits_uqc_full_product_state_variance).rolling(100).mean()
four_qubits_uqc_partial_product_state_variance_smoothed = pd.Series(four_qubits_uqc_partial_product_state_variance).rolling(100).mean()
skolik_datareup_variance_smoothed = pd.Series(skolik_datareup_variance).rolling(100).mean()
````

````python
fig,axs = plt.subplots(1,1,figsize=(7.5,5), tight_layout=True)
axs.plot(skolik_datareup_mean_rewards, color = "blue", label="Skolik Data Re-Uploading")
axs.plot(one_qubit_uqc_mean_rewards, color = "red", label="Single-qubit UQC")
axs.fill_between(range(len(skolik_datareup_mean_rewards)), np.clip(skolik_datareup_mean_rewards - skolik_datareup_std_rewards,a_min = 0, a_max=200), np.clip(skolik_datareup_mean_rewards + skolik_datareup_std_rewards,a_min = 0, a_max=200), alpha=0.2, color = "blue")
axs.fill_between(range(len(one_qubit_uqc_mean_rewards)), np.clip(one_qubit_uqc_mean_rewards - one_qubit_uqc_std_rewards,a_min = 0, a_max=200), np.clip(one_qubit_uqc_mean_rewards + one_qubit_uqc_std_rewards,a_min = 0, a_max=200), alpha=0.2, color = "red")
axs.set_xlabel("Episode")
axs.set_ylabel("Mean Return")
axs.grid(True)
axs.legend()

plt.show()
````
