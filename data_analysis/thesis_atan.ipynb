```python
import numpy as np
import matplotlib.pyplot as plt
import pandas as pd
from analysis_functions import Analysis
```

```python

skolik_datareup_path = "../../results/thesis/1.1/skolik_datareup"
skolik_datareup_atan_path = "../../results/thesis_arctan/skolik_datareup"

skolik_datareup = Analysis(skolik_datareup_path)
skolik_datareup_atan = Analysis(skolik_datareup_atan_path)
```

```python
skolik_datareup_return = skolik_datareup.get_rewards()
skolik_datareup_atan_return = skolik_datareup_atan.get_rewards()

skolik_datareup_mean_return = np.mean(skolik_datareup_return, axis=0)
skolik_datareup_atan_mean_return = np.mean(skolik_datareup_atan_return, axis=0)

skolik_datareup_std_return = np.std(skolik_datareup_return, axis=0)
skolik_datareup_atan_std_return = np.std(skolik_datareup_atan_return, axis=0)
```

```python
plt.plot(skolik_datareup_mean_return, label="skolik_datareup")
plt.plot(skolik_datareup_atan_mean_return, label="skolik_datareup_atan")
plt.fill_between(np.arange(len(skolik_datareup_mean_return)), skolik_datareup_mean_return - skolik_datareup_std_return, skolik_datareup_mean_return + skolik_datareup_std_return, alpha=0.2)
plt.fill_between(np.arange(len(skolik_datareup_atan_mean_return)), skolik_datareup_atan_mean_return - skolik_datareup_atan_std_return, skolik_datareup_atan_mean_return + skolik_datareup_atan_std_return, alpha=0.2)
```
