# 1.1) Skolik Circuit with and without Data Re-Uploading and with or without scalings

## 1.1.1) Performance

```python
skolik_baseline_return = skolik_baseline.get_rewards()
skolik_datareup_return = skolik_datareup.get_rewards()
skolik_baseline_no_in_return = skolik_baseline_no_in.get_rewards()
skolik_datareup_no_in_return = skolik_datareup_no_in.get_rewards()
skolik_baseline_no_out_return = skolik_baseline_no_out.get_rewards()
skolik_datareup_no_out_return = skolik_datareup_no_out.get_rewards()
skolik_baseline_no_in_out_return = skolik_baseline_no_in_out.get_rewards()
skolik_datareup_no_in_out_return = skolik_datareup_no_in_out.get_rewards()

skolik_baseline_return_mean = np.mean(skolik_baseline_return, axis=0)
skolik_datareup_return_mean = np.mean(skolik_datareup_return, axis=0)
skolik_baseline_no_in_return_mean = np.mean(skolik_baseline_no_in_return, axis=0)
skolik_datareup_no_in_return_mean = np.mean(skolik_datareup_no_in_return, axis=0)
skolik_baseline_no_out_return_mean = np.mean(skolik_baseline_no_out_return, axis=0)
skolik_datareup_no_out_return_mean = np.mean(skolik_datareup_no_out_return, axis=0)
skolik_baseline_no_in_out_return_mean = np.mean(skolik_baseline_no_in_out_return, axis=0)
skolik_datareup_no_in_out_return_mean = np.mean(skolik_datareup_no_in_out_return, axis=0)

skolik_baseline_return_std = np.std(skolik_baseline_return, axis=0)
skolik_datareup_return_std = np.std(skolik_datareup_return, axis=0)
skolik_baseline_no_in_return_std = np.std(skolik_baseline_no_in_return, axis=0)
skolik_datareup_no_in_return_std = np.std(skolik_datareup_no_in_return, axis=0)
skolik_baseline_no_out_return_std = np.std(skolik_baseline_no_out_return, axis=0)
skolik_datareup_no_out_return_std = np.std(skolik_datareup_no_out_return, axis=0)
skolik_baseline_no_in_out_return_std = np.std(skolik_baseline_no_in_out_return, axis=0)
skolik_datareup_no_in_out_return_std = np.std(skolik_datareup_no_in_out_return, axis=0)
```

```python
fig,axs = plt.subplots(1,3,figsize=(15,5), tight_layout=True)

axs[0].plot(skolik_baseline_return_mean,color = "red", label="Baseline")
axs[0].plot(skolik_datareup_return_mean,color = "blue", label="Data Reuploading")
axs[0].fill_between(np.arange(len(skolik_baseline_return_mean)), skolik_baseline_return_mean - skolik_baseline_return_std, np.clip(skolik_baseline_return_mean + skolik_baseline_return_std, a_min = None, a_max = 200), alpha=0.2, color="red")
axs[0].fill_between(np.arange(len(skolik_datareup_return_mean)), skolik_datareup_return_mean - skolik_datareup_return_std, np.clip(skolik_datareup_return_mean + skolik_datareup_return_std, a_min = None, a_max = 200), alpha=0.2, color="blue")
axs[0].set_title("Baseline vs Data Reuploading")
axs[0].set_xlabel("Episode")
axs[0].set_ylabel("Mean Return")
axs[0].legend()

axs[1].plot(skolik_baseline_return_mean,color = "red", label="Baseline")
axs[1].plot(skolik_baseline_no_in_return_mean,color = "green", label="Baseline w/o input scaling")
axs[1].plot(skolik_baseline_no_out_return_mean,color = "orange", label="Baseline w/o output scaling")
axs[1].plot(skolik_baseline_no_in_out_return_mean,color = "purple", label="Baseline w/o scaling")
axs[1].fill_between(np.arange(len(skolik_baseline_return_mean)), skolik_baseline_return_mean - skolik_baseline_return_std, np.clip(skolik_baseline_return_mean + skolik_baseline_return_std, a_min = None, a_max = 200), alpha=0.2, color="red")
axs[1].fill_between(np.arange(len(skolik_baseline_no_in_return_mean)), skolik_baseline_no_in_return_mean - skolik_baseline_no_in_return_std, np.clip(skolik_baseline_no_in_return_mean + skolik_baseline_no_in_return_std, a_min = None, a_max = 200), alpha=0.2, color="green")
axs[1].fill_between(np.arange(len(skolik_baseline_no_out_return_mean)), skolik_baseline_no_out_return_mean - skolik_baseline_no_out_return_std, np.clip(skolik_baseline_no_out_return_mean + skolik_baseline_no_out_return_std, a_min = None, a_max = 200), alpha=0.2, color="orange")
axs[1].fill_between(np.arange(len(skolik_baseline_no_in_out_return_mean)), skolik_baseline_no_in_out_return_mean - skolik_baseline_no_in_out_return_std, np.clip(skolik_baseline_no_in_out_return_mean + skolik_baseline_no_in_out_return_std, a_min = None, a_max = 200), alpha=0.2, color="purple")
axs[1].set_title("Baseline with and without scaling")
axs[1].set_xlabel("Episode")
axs[1].set_ylabel("Mean Return")
axs[1].legend()

axs[2].plot(skolik_datareup_return_mean,color = "blue", label="Data Reuploading")
axs[2].plot(skolik_datareup_no_in_return_mean,color = "green", label="Data Reuploading w/o input scaling")
axs[2].plot(skolik_datareup_no_out_return_mean,color = "orange", label="Data Reuploading w/o output scaling")
axs[2].plot(skolik_datareup_no_in_out_return_mean,color = "purple", label="Data Reuploading w/o scaling")
axs[2].fill_between(np.arange(len(skolik_datareup_return_mean)), skolik_datareup_return_mean - skolik_datareup_return_std, np.clip(skolik_datareup_return_mean + skolik_datareup_return_std, a_min = None, a_max = 200), alpha=0.2, color="blue")
axs[2].fill_between(np.arange(len(skolik_datareup_no_in_return_mean)), skolik_datareup_no_in_return_mean - skolik_datareup_no_in_return_std, np.clip(skolik_datareup_no_in_return_mean + skolik_datareup_no_in_return_std, a_min = None, a_max = 200), alpha=0.2, color="green")
axs[2].fill_between(np.arange(len(skolik_datareup_no_out_return_mean)), skolik_datareup_no_out_return_mean - skolik_datareup_no_out_return_std, np.clip(skolik_datareup_no_out_return_mean + skolik_datareup_no_out_return_std, a_min = None, a_max = 200), alpha=0.2, color="orange")
axs[2].fill_between(np.arange(len(skolik_datareup_no_in_out_return_mean)), skolik_datareup_no_in_out_return_mean - skolik_datareup_no_in_out_return_std, np.clip(skolik_datareup_no_in_out_return_mean + skolik_datareup_no_in_out_return_std, a_min = None, a_max = 200), alpha=0.2, color="purple")
axs[2].set_title("Data Reuploading with and without scaling")
axs[2].set_xlabel("Episode")
axs[2].set_ylabel("Mean Return")
axs[2].legend()

plt.show()
```

# 1.2) The UQC architecture

```python
uqc_1qubit_XY_continuous_path = "../../results/thesis_organized/1.2/uqc_1qubit_XY_continuous"
uqc_1qubit_ZX_continuous_path = "../../results/thesis_organized/1.2/uqc_1qubit_ZX_continuous"
uqc_1qubit_ZY_continuous_path = "../../results/thesis_organized/1.2/uqc_1qubit_ZY_continuous"

uqc_1qubit_XY_nothing_path = "../../results/thesis_organized/1.2/uqc_1qubit_XY_nothing"
uqc_1qubit_ZX_nothing_path = "../../results/thesis_organized/1.2/uqc_1qubit_ZX_nothing"
uqc_1qubit_ZY_nothing_path = "../../results/thesis_organized/1.2/uqc_1qubit_ZY_nothing"

uqc_1qubit_XY_continuous = Analysis(uqc_1qubit_XY_continuous_path)
uqc_1qubit_ZX_continuous = Analysis(uqc_1qubit_ZX_continuous_path)
uqc_1qubit_ZY_continuous = Analysis(uqc_1qubit_ZY_continuous_path)

uqc_1qubit_XY_nothing = Analysis(uqc_1qubit_XY_nothing_path)
uqc_1qubit_ZX_nothing = Analysis(uqc_1qubit_ZX_nothing_path)
uqc_1qubit_ZY_nothing = Analysis(uqc_1qubit_ZY_nothing_path)
```

## 1.2.1) Performance

```python
uqc_1qubit_XY_nothing_return = uqc_1qubit_XY_nothing.get_rewards()
uqc_1qubit_ZX_nothing_return = uqc_1qubit_ZX_nothing.get_rewards()
uqc_1qubit_ZY_nothing_return = uqc_1qubit_ZY_nothing.get_rewards()

uqc_1qubit_XY_nothing_mean = np.mean(uqc_1qubit_XY_nothing_return,axis=0)
uqc_1qubit_ZX_nothing_mean = np.mean(uqc_1qubit_ZX_nothing_return,axis=0)
uqc_1qubit_ZY_nothing_mean = np.mean(uqc_1qubit_ZY_nothing_return,axis=0)

uqc_1qubit_XY_nothing_std = np.std(uqc_1qubit_XY_nothing_return,axis=0)
uqc_1qubit_ZX_nothing_std = np.std(uqc_1qubit_ZX_nothing_return,axis=0)
uqc_1qubit_ZY_nothing_std = np.std(uqc_1qubit_ZY_nothing_return,axis=0)

uqc_1qubit_XY_continuous_return = uqc_1qubit_XY_continuous.get_rewards()
uqc_1qubit_ZX_continuous_return = uqc_1qubit_ZX_continuous.get_rewards()
uqc_1qubit_ZY_continuous_return = uqc_1qubit_ZY_continuous.get_rewards()

uqc_1qubit_XY_continuous_mean = np.mean(uqc_1qubit_XY_continuous_return,axis=0)
uqc_1qubit_ZX_continuous_mean = np.mean(uqc_1qubit_ZX_continuous_return,axis=0)
uqc_1qubit_ZY_continuous_mean = np.mean(uqc_1qubit_ZY_continuous_return,axis=0)

uqc_1qubit_XY_continuous_std = np.std(uqc_1qubit_XY_continuous_return,axis=0)
uqc_1qubit_ZX_continuous_std = np.std(uqc_1qubit_ZX_continuous_return,axis=0)
uqc_1qubit_ZY_continuous_std = np.std(uqc_1qubit_ZY_continuous_return,axis=0)
```

```python
fig, axs = plt.subplots(1,2,figsize=(15,5), tight_layout=True)

axs[0].plot(uqc_1qubit_XY_nothing_mean,color = "red", label="XY")
axs[0].fill_between(np.arange(len(uqc_1qubit_XY_nothing_mean)),uqc_1qubit_XY_nothing_mean-uqc_1qubit_XY_nothing_std,np.clip(uqc_1qubit_XY_nothing_mean+uqc_1qubit_XY_nothing_std, a_min = None, a_max = 200),alpha=0.2,color="red")
axs[0].plot(uqc_1qubit_ZX_nothing_mean,color = "blue", label="ZX")
axs[0].fill_between(np.arange(len(uqc_1qubit_ZX_nothing_mean)),uqc_1qubit_ZX_nothing_mean-uqc_1qubit_ZX_nothing_std,np.clip(uqc_1qubit_ZX_nothing_mean+uqc_1qubit_ZX_nothing_std, a_min = None, a_max = 200),alpha=0.2,color="blue")
axs[0].plot(uqc_1qubit_ZY_nothing_mean,color = "green", label="ZY")
axs[0].fill_between(np.arange(len(uqc_1qubit_ZY_nothing_mean)),uqc_1qubit_ZY_nothing_mean-uqc_1qubit_ZY_nothing_std,np.clip(uqc_1qubit_ZY_nothing_mean+uqc_1qubit_ZY_nothing_std, a_min = None, a_max = 200),alpha=0.2,color="green")
axs[0].set_title("UQC 1 Qubit Nothing Encoding")
axs[0].set_xlabel("Episode")
axs[0].set_ylabel("Average Return")
axs[0].legend()

axs[1].plot(uqc_1qubit_XY_continuous_mean,color = "red", label="XY")
axs[1].fill_between(np.arange(len(uqc_1qubit_XY_continuous_mean)),uqc_1qubit_XY_continuous_mean-uqc_1qubit_XY_continuous_std,np.clip(uqc_1qubit_XY_continuous_mean+uqc_1qubit_XY_continuous_std, a_min = None, a_max = 200),alpha=0.2,color="red")
axs[1].plot(uqc_1qubit_ZX_continuous_mean,color = "blue", label="ZX")
axs[1].fill_between(np.arange(len(uqc_1qubit_ZX_continuous_mean)),uqc_1qubit_ZX_continuous_mean-uqc_1qubit_ZX_continuous_std,np.clip(uqc_1qubit_ZX_continuous_mean+uqc_1qubit_ZX_continuous_std, a_min = None, a_max = 200),alpha=0.2,color="blue")
axs[1].plot(uqc_1qubit_ZY_continuous_mean,color = "green", label="ZY")
axs[1].fill_between(np.arange(len(uqc_1qubit_ZY_continuous_mean)),uqc_1qubit_ZY_continuous_mean-uqc_1qubit_ZY_continuous_std,np.clip(uqc_1qubit_ZY_continuous_mean+uqc_1qubit_ZY_continuous_std, a_min = None, a_max = 200),alpha=0.2,color="green")
axs[1].set_title("UQC 1 Qubit Continuous Encoding")
axs[1].set_xlabel("Episode")
axs[1].set_ylabel("Average Return")
axs[1].legend()

plt.show()
```
