```python
import numpy as np
import matplotlib.pyplot as plt
import pandas as pd
from analysis_functions import Analysis
```

```python
skolik_baseline_path = "../../results/thesis_organized/1.1/skolik_baseline"
skolik_datareup_path = "../../results/thesis_organized/1.1/skolik_datareup"
skolik_baseline_no_in_path = "../../results/thesis_organized/1.1/skolik_baseline_no_in"
skolik_datareup_no_in_path = "../../results/thesis_organized/1.1/skolik_datareup_no_in"
skolik_baseline_no_out_path = "../../results/thesis_organized/1.1/skolik_baseline_no_out"
skolik_datareup_no_out_path = "../../results/thesis_organized/1.1/skolik_datareup_no_out"
skolik_baseline_no_in_out_path = "../../results/thesis_organized/1.1/skolik_baseline_no_in_out"
skolik_datareup_no_in_out_path = "../../results/thesis_organized/1.1/skolik_datareup_no_in_out"

skolik_baseline = Analysis(skolik_baseline_path)
skolik_datareup = Analysis(skolik_datareup_path)
skolik_baseline_no_in = Analysis(skolik_baseline_no_in_path)
skolik_datareup_no_in = Analysis(skolik_datareup_no_in_path)
skolik_baseline_no_out = Analysis(skolik_baseline_no_out_path)
skolik_datareup_no_out = Analysis(skolik_datareup_no_out_path)
skolik_baseline_no_in_out = Analysis(skolik_baseline_no_in_out_path)
skolik_datareup_no_in_out = Analysis(skolik_datareup_no_in_out_path)
```

# 1.1) Skolik Circuit with and without Data Re-Uploading and with or without scalings

## 1.1.1) Performance

```python
skolik_baseline_return = skolik_baseline.get_rewards()
skolik_datareup_return = skolik_datareup.get_rewards()
skolik_baseline_no_in_return = skolik_baseline_no_in.get_rewards()
skolik_datareup_no_in_return = skolik_datareup_no_in.get_rewards()
skolik_baseline_no_out_return = skolik_baseline_no_out.get_rewards()
skolik_datareup_no_out_return = skolik_datareup_no_out.get_rewards()
skolik_baseline_no_in_out_return = skolik_baseline_no_in_out.get_rewards()
skolik_datareup_no_in_out_return = skolik_datareup_no_in_out.get_rewards()

skolik_baseline_return_mean = np.mean(skolik_baseline_return, axis=0)
skolik_datareup_return_mean = np.mean(skolik_datareup_return, axis=0)
skolik_baseline_no_in_return_mean = np.mean(skolik_baseline_no_in_return, axis=0)
skolik_datareup_no_in_return_mean = np.mean(skolik_datareup_no_in_return, axis=0)
skolik_baseline_no_out_return_mean = np.mean(skolik_baseline_no_out_return, axis=0)
skolik_datareup_no_out_return_mean = np.mean(skolik_datareup_no_out_return, axis=0)
skolik_baseline_no_in_out_return_mean = np.mean(skolik_baseline_no_in_out_return, axis=0)
skolik_datareup_no_in_out_return_mean = np.mean(skolik_datareup_no_in_out_return, axis=0)

skolik_baseline_return_std = np.std(skolik_baseline_return, axis=0)
skolik_datareup_return_std = np.std(skolik_datareup_return, axis=0)
skolik_baseline_no_in_return_std = np.std(skolik_baseline_no_in_return, axis=0)
skolik_datareup_no_in_return_std = np.std(skolik_datareup_no_in_return, axis=0)
skolik_baseline_no_out_return_std = np.std(skolik_baseline_no_out_return, axis=0)
skolik_datareup_no_out_return_std = np.std(skolik_datareup_no_out_return, axis=0)
skolik_baseline_no_in_out_return_std = np.std(skolik_baseline_no_in_out_return, axis=0)
skolik_datareup_no_in_out_return_std = np.std(skolik_datareup_no_in_out_return, axis=0)
```

```python
fig,axs = plt.subplots(1,3,figsize=(15,5), tight_layout=True)

axs[0].plot(skolik_baseline_return_mean,color = "red", label="Baseline")
axs[0].plot(skolik_datareup_return_mean,color = "blue", label="Data Reuploading")
axs[0].fill_between(np.arange(len(skolik_baseline_return_mean)), skolik_baseline_return_mean - skolik_baseline_return_std, skolik_baseline_return_mean + skolik_baseline_return_std, alpha=0.2, color="red")
axs[0].fill_between(np.arange(len(skolik_datareup_return_mean)), skolik_datareup_return_mean - skolik_datareup_return_std, skolik_datareup_return_mean + skolik_datareup_return_std, alpha=0.2, color="blue")
axs[0].set_title("Baseline vs Data Reuploading")
axs[0].set_xlabel("Episode")
axs[0].set_ylabel("Mean Return")
axs[0].legend()

axs[1].plot(skolik_baseline_return_mean,color = "red", label="Baseline")
axs[1].plot(skolik_baseline_no_in_return_mean,color = "green", label="Baseline w/o input scaling")
axs[1].plot(skolik_baseline_no_out_return_mean,color = "orange", label="Baseline w/o output scaling")
axs[1].plot(skolik_baseline_no_in_out_return_mean,color = "purple", label="Baseline w/o scaling")
axs[1].fill_between(np.arange(len(skolik_baseline_return_mean)), skolik_baseline_return_mean - skolik_baseline_return_std, skolik_baseline_return_mean + skolik_baseline_return_std, alpha=0.2, color="red")
axs[1].fill_between(np.arange(len(skolik_baseline_no_in_return_mean)), skolik_baseline_no_in_return_mean - skolik_baseline_no_in_return_std, skolik_baseline_no_in_return_mean + skolik_baseline_no_in_return_std, alpha=0.2, color="green")
axs[1].fill_between(np.arange(len(skolik_baseline_no_out_return_mean)), skolik_baseline_no_out_return_mean - skolik_baseline_no_out_return_std, skolik_baseline_no_out_return_mean + skolik_baseline_no_out_return_std, alpha=0.2, color="orange")
axs[1].fill_between(np.arange(len(skolik_baseline_no_in_out_return_mean)), skolik_baseline_no_in_out_return_mean - skolik_baseline_no_in_out_return_std, skolik_baseline_no_in_out_return_mean + skolik_baseline_no_in_out_return_std, alpha=0.2, color="purple")
axs[1].set_title("Baseline with and without scaling")
axs[1].set_xlabel("Episode")
axs[1].set_ylabel("Mean Return")
axs[1].legend()

axs[2].plot(skolik_datareup_return_mean,color = "blue", label="Data Reuploading")
axs[2].plot(skolik_datareup_no_in_return_mean,color = "green", label="Data Reuploading w/o input scaling")
axs[2].plot(skolik_datareup_no_out_return_mean,color = "orange", label="Data Reuploading w/o output scaling")
axs[2].plot(skolik_datareup_no_in_out_return_mean,color = "purple", label="Data Reuploading w/o scaling")
axs[2].fill_between(np.arange(len(skolik_datareup_return_mean)), skolik_datareup_return_mean - skolik_datareup_return_std, skolik_datareup_return_mean + skolik_datareup_return_std, alpha=0.2, color="blue")
axs[2].fill_between(np.arange(len(skolik_datareup_no_in_return_mean)), skolik_datareup_no_in_return_mean - skolik_datareup_no_in_return_std, skolik_datareup_no_in_return_mean + skolik_datareup_no_in_return_std, alpha=0.2, color="green")
axs[2].fill_between(np.arange(len(skolik_datareup_no_out_return_mean)), skolik_datareup_no_out_return_mean - skolik_datareup_no_out_return_std, skolik_datareup_no_out_return_mean + skolik_datareup_no_out_return_std, alpha=0.2, color="orange")
axs[2].fill_between(np.arange(len(skolik_datareup_no_in_out_return_mean)), skolik_datareup_no_in_out_return_mean - skolik_datareup_no_in_out_return_std, skolik_datareup_no_in_out_return_mean + skolik_datareup_no_in_out_return_std, alpha=0.2, color="purple")
axs[2].set_title("Data Reuploading with and without scaling")
axs[2].set_xlabel("Episode")
axs[2].set_ylabel("Mean Return")
axs[2].legend()

plt.show()
```

## 1.1.2) Gradients

```python
skolik_baseline_mean_gradients, skolik_baseline_variance_gradients = skolik_baseline.calculate_mean_variance_gradients()
skolik_baseline_no_in_mean_gradients, skolik_baseline_no_in_variance_gradients = skolik_baseline_no_in.calculate_mean_variance_gradients()
skolik_baseline_no_out_mean_gradients, skolik_baseline_no_out_variance_gradients = skolik_baseline_no_out.calculate_mean_variance_gradients()
skolik_baseline_no_in_out_mean_gradients, skolik_baseline_no_in_out_variance_gradients = skolik_baseline_no_in_out.calculate_mean_variance_gradients()
skolik_datareup_mean_gradients, skolik_datareup_variance_gradients = skolik_datareup.calculate_mean_variance_gradients()
skolik_datareup_no_in_mean_gradients, skolik_datareup_no_in_variance_gradients = skolik_datareup_no_in.calculate_mean_variance_gradients()
skolik_datareup_no_out_mean_gradients, skolik_datareup_no_out_variance_gradients = skolik_datareup_no_out.calculate_mean_variance_gradients()
skolik_datareup_no_in_out_mean_gradients, skolik_datareup_no_in_out_variance_gradients = skolik_datareup_no_in_out.calculate_mean_variance_gradients()
```

```python
skolik_baseline_mean_gradients_smoothed = pd.Series(skolik_baseline_mean_gradients).rolling(100).mean()
skolik_baseline_no_in_mean_gradients_smoothed = pd.Series(skolik_baseline_no_in_mean_gradients).rolling(100).mean()
skolik_baseline_no_out_mean_gradients_smoothed = pd.Series(skolik_baseline_no_out_mean_gradients).rolling(100).mean()
skolik_baseline_no_in_out_mean_gradients_smoothed = pd.Series(skolik_baseline_no_in_out_mean_gradients).rolling(100).mean()
skolik_datareup_mean_gradients_smoothed = pd.Series(skolik_datareup_mean_gradients).rolling(100).mean()
skolik_datareup_no_in_mean_gradients_smoothed = pd.Series(skolik_datareup_no_in_mean_gradients).rolling(100).mean()
skolik_datareup_no_out_mean_gradients_smoothed = pd.Series(skolik_datareup_no_out_mean_gradients).rolling(100).mean()
skolik_datareup_no_in_out_mean_gradients_smoothed = pd.Series(skolik_datareup_no_in_out_mean_gradients).rolling(100).mean()

skolik_baseline_variance_gradients_smoothed = pd.Series(skolik_baseline_variance_gradients).rolling(100).mean()
skolik_baseline_no_in_variance_gradients_smoothed = pd.Series(skolik_baseline_no_in_variance_gradients).rolling(100).mean()
skolik_baseline_no_out_variance_gradients_smoothed = pd.Series(skolik_baseline_no_out_variance_gradients).rolling(100).mean()
skolik_baseline_no_in_out_variance_gradients_smoothed = pd.Series(skolik_baseline_no_in_out_variance_gradients).rolling(100).mean()
skolik_datareup_variance_gradients_smoothed = pd.Series(skolik_datareup_variance_gradients).rolling(100).mean()
skolik_datareup_no_in_variance_gradients_smoothed = pd.Series(skolik_datareup_no_in_variance_gradients).rolling(100).mean()
skolik_datareup_no_out_variance_gradients_smoothed = pd.Series(skolik_datareup_no_out_variance_gradients).rolling(100).mean()
skolik_datareup_no_in_out_variance_gradients_smoothed = pd.Series(skolik_datareup_no_in_out_variance_gradients).rolling(100).mean()
```

```python
fig, axs = plt.subplots(2, 3, figsize=(15,8.5), tight_layout=True)

axs[0,0].plot(skolik_baseline_mean_gradients_smoothed,color = "red", label="Baseline")
axs[0,0].plot(skolik_datareup_mean_gradients_smoothed,color = "blue", label="Data  ReUploading")
axs[0,0].set_title("Baseline Vs DataReUploading")
axs[0,0].set_xlabel("Training Step")
axs[0,0].set_ylabel("Mean Magnitude of Gradient Vector")
axs[0,0].legend()

axs[0,1].plot(skolik_baseline_mean_gradients_smoothed,color = "red", label="Baseline")
axs[0,1].plot(skolik_baseline_no_in_mean_gradients_smoothed,color = "green", label="Baseline w/o input scaling")
axs[0,1].plot(skolik_baseline_no_out_mean_gradients_smoothed,color = "orange", label="Baseline w/o output scaling")
axs[0,1].plot(skolik_baseline_no_in_out_mean_gradients_smoothed,color = "purple", label="Baseline w/o scaling")
axs[0,1].set_title("Baseline with and without scaling")
axs[0,1].set_xlabel("Training Step")
axs[0,1].set_ylabel("Mean Magnitude of Gradient Vector")
axs[0,1].legend()

axs[0,2].plot(skolik_datareup_mean_gradients_smoothed,color = "blue", label="Data ReUploading")
axs[0,2].plot(skolik_datareup_no_in_mean_gradients_smoothed,color = "green", label="Data ReUploading w/o input scaling")
axs[0,2].plot(skolik_datareup_no_out_mean_gradients_smoothed,color = "orange", label="Data ReUploading w/o output scaling")
axs[0,2].plot(skolik_datareup_no_in_out_mean_gradients_smoothed,color = "purple", label="Data ReUploading w/o scaling")
axs[0,2].set_title("Data ReUploading with and without scaling")
axs[0,2].set_xlabel("Training Step")
axs[0,2].set_ylabel("Mean Magnitude of Gradient Vector")
axs[0,2].legend()

axs[1,0].plot(skolik_baseline_variance_gradients_smoothed,color = "red", label="Baseline")
axs[1,0].plot(skolik_datareup_variance_gradients_smoothed,color = "blue", label="Data  ReUploading")
axs[1,0].set_title("Baseline Vs DataReUploading")
axs[1,0].set_xlabel("Training Step")
axs[1,0].set_ylabel("Variance of the Magnitude of the Gradient Vector")
axs[1,0].legend()

axs[1,1].plot(skolik_baseline_variance_gradients_smoothed,color = "red", label="Baseline")
axs[1,1].plot(skolik_baseline_no_in_variance_gradients_smoothed,color = "green", label="Baseline w/o input scaling")
axs[1,1].plot(skolik_baseline_no_out_variance_gradients_smoothed,color = "orange", label="Baseline w/o output scaling")
axs[1,1].plot(skolik_baseline_no_in_out_variance_gradients_smoothed,color = "purple", label="Baseline w/o scaling")
axs[1,1].set_title("Baseline with and without scaling")
axs[1,1].set_xlabel("Training Step")
axs[1,1].set_ylabel("Variance of the Magnitude of the Gradient Vector")
axs[1,1].legend()

axs[1,2].plot(skolik_datareup_variance_gradients_smoothed,color = "blue", label="Data ReUploading")
axs[1,2].plot(skolik_datareup_no_in_variance_gradients_smoothed,color = "green", label="Data ReUploading w/o input scaling")
axs[1,2].plot(skolik_datareup_no_out_variance_gradients_smoothed,color = "orange", label="Data ReUploading w/o output scaling")
axs[1,2].plot(skolik_datareup_no_in_out_variance_gradients_smoothed,color = "purple", label="Data ReUploading w/o scaling")
axs[1,2].set_title("Data ReUploading with and without scaling")
axs[1,2].set_xlabel("Training Step")
axs[1,2].set_ylabel("Variance of the Magnitude of the Gradient Vector")
axs[1,2].legend()

plt.show()
```

# 1.2) The UQC architecture

```python
uqc_1qubit_XY_path = "../../results/thesis_organized/1.2/uqc_1qubit_XY"
uqc_1qubit_ZX_path = "../../results/thesis_organized/1.2/uqc_1qubit_ZX"
uqc_1qubit_ZY_path = "../../results/thesis_organized/1.2/uqc_1qubit_ZY"

uqc_1qubit_XY = Analysis(uqc_1qubit_XY_path)
uqc_1qubit_ZX = Analysis(uqc_1qubit_ZX_path)
uqc_1qubit_ZY = Analysis(uqc_1qubit_ZY_path)
```

## 1.2.1) Performance

```python
uqc_1qubit_XY_return = uqc_1qubit_XY.get_rewards()
uqc_1qubit_ZX_return = uqc_1qubit_ZX.get_rewards()
uqc_1qubit_ZY_return = uqc_1qubit_ZY.get_rewards()

uqc_1qubit_XY_mean = np.mean(uqc_1qubit_XY_return,axis=0)
uqc_1qubit_ZX_mean = np.mean(uqc_1qubit_ZX_return,axis=0)
uqc_1qubit_ZY_mean = np.mean(uqc_1qubit_ZY_return,axis=0)

uqc_1qubit_XY_std = np.std(uqc_1qubit_XY_return,axis=0)
uqc_1qubit_ZX_std = np.std(uqc_1qubit_ZX_return,axis=0)
uqc_1qubit_ZY_std = np.std(uqc_1qubit_ZY_return,axis=0)
```

```python
plt.plot(uqc_1qubit_XY_mean,color="red",label="XY")
plt.fill_between(np.arange(len(uqc_1qubit_XY_mean)),uqc_1qubit_XY_mean-uqc_1qubit_XY_std,uqc_1qubit_XY_mean+uqc_1qubit_XY_std,alpha=0.2,color="red")
plt.plot(uqc_1qubit_ZX_mean,color="blue",label="ZX")
plt.fill_between(np.arange(len(uqc_1qubit_ZX_mean)),uqc_1qubit_ZX_mean-uqc_1qubit_ZX_std,uqc_1qubit_ZX_mean+uqc_1qubit_ZX_std,alpha=0.2,color="blue")
plt.plot(uqc_1qubit_ZY_mean,color="green",label="ZY")
plt.fill_between(np.arange(len(uqc_1qubit_ZY_mean)),uqc_1qubit_ZY_mean-uqc_1qubit_ZY_std,uqc_1qubit_ZY_mean+uqc_1qubit_ZY_std,alpha=0.2,color="green")
plt.title("UQC 1 Qubit")
plt.xlabel("Episode")
plt.ylabel("Mean Return")
plt.legend()
plt.show()
```
