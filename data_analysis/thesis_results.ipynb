```python
import numpy as np
from scipy.optimize import curve_fit

# Given data
y_values = np.asarray([mean_variance_1qubit, mean_variance_2qubits, mean_variance_4qubits])
x_values = np.asarray([1,2,4], dtype=np.float64) # Assuming the x values are [0, 1, 2] for the given y values

# Define the exponential function to fit the data
def exponential_func(x, a, b):
    return a * np.exp(-b * x)

# Fit the data to the exponential function using curve_fit
popt, _ = curve_fit(exponential_func, x_values, y_values)

# Get the fitted parameters
a_fit, b_fit = popt

# Calculate the R-squared value
y_fit = exponential_func(x_values, a_fit, b_fit)
residuals = y_values - y_fit
ss_res = np.sum(residuals**2)
ss_tot = np.sum((y_values - np.mean(y_values))**2)
r_squared = 1 - (ss_res / ss_tot)

# Check if the decline is exponential based on the R-squared value
threshold_r_squared = 0.95  #
if r_squared >= threshold_r_squared:
    print("The decline appears to be exponential.")
else:
    print("The decline does not appear to be exponential.")

# Plot the original data points
plt.scatter(x_values, y_values, label="Original Data")

# Plot the fitted exponential curve
plt.plot(x_values, y_fit, 'r-', label="Fitted Exponential Curve")

# Annotate the R-squared value on the plot
plt.text(0.5, 0.6, f"R-squared: {r_squared:.3f}", transform=plt.gca().transAxes, fontsize=12)

# Set plot labels and title
plt.xlabel("X Values")
plt.ylabel("Y Values")
plt.yscale("log")
plt.title("Exponential Fit to Data")
plt.legend()

# Display the plot
plt.show()
```

```python
import numpy as np
import matplotlib.pyplot as plt
from scipy.optimize import curve_fit

# Given data
y_values = np.asarray([mean_variance_1qubit, mean_variance_2qubits, mean_variance_4qubits])
x_values = np.asarray([1,2,4], dtype = np.float64)  # Assuming the x values are [0, 1, 2] for the given y values

# Define the polynomial function to fit the data
def polynomial_func(x, *coefficients):
    return np.polyval(coefficients, x)

initial_guess = [1, -1, 1]

# Fit the data to the polynomial function using curve_fit
popt, _ = curve_fit(polynomial_func, x_values, y_values, p0=initial_guess)

# Calculate the R-squared value for polynomial fit
y_fit = polynomial_func(x_values, *popt)
residuals = y_values - y_fit
ss_res = np.sum(residuals**2)
ss_tot = np.sum((y_values - np.mean(y_values))**2)
r_squared = 1 - (ss_res / ss_tot)

# Check if the data is polynomial based on the R-squared value
threshold_r_squared = 0.95  # You can adjust this threshold based on your preference
if r_squared >= threshold_r_squared:
    print("The data appears to follow a polynomial pattern.")
else:
    print("The data does not appear to follow a polynomial pattern.")

# Plot the original data points
plt.scatter(x_values, y_values, label="Original Data")

# Plot the fitted polynomial curve
x_fit = np.linspace(0, len(y_values) - 1, 100)  # Generate x values for smooth curve plotting
y_fit = polynomial_func(x_fit, *popt)
plt.plot(x_fit, y_fit, 'r-', label=f"Fitted Polynomial")

# Annotate the R-squared value on the plot
plt.text(0.5, 0.9, f"R-squared: {r_squared:.3f}", transform=plt.gca().transAxes, fontsize=12)

# Set plot labels and title
plt.xlabel("X Values")
plt.ylabel("Y Values")
plt.title("Polynomial Fit to Data")
plt.legend()

# Display the plot
plt.show()
```
