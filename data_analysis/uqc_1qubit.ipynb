```python
import numpy as np
from analysis_functions import Analysis
import matplotlib.pyplot as plt
import pandas as pd
```

```python
uqc_1qubit_XY_continuous_path = "../../results/uqc_test/uqc_1qubit_XY_continuous"
uqc_1qubit_XY_nothing_path = "../../results/uqc_test/uqc_1qubit_XY_nothing"
uqc_1qubit_ZX_continuous_path = "../../results/uqc_test/uqc_1qubit_ZX_continuous"
uqc_1qubit_ZX_nothing_path = "../../results/uqc_test/uqc_1qubit_ZX_nothing"
uqc_1qubit_ZY_continuous_path = "../../results/uqc_test/uqc_1qubit_ZY_continuous"
uqc_1qubit_ZY_nothing_path = "../../results/uqc_test/uqc_1qubit_ZY_nothing"

uqc_1qubit_XY_continuous = Analysis(uqc_1qubit_XY_continuous_path)
uqc_1qubit_XY_nothing = Analysis(uqc_1qubit_XY_nothing_path)
uqc_1qubit_ZX_continuous = Analysis(uqc_1qubit_ZX_continuous_path)
uqc_1qubit_ZX_nothing = Analysis(uqc_1qubit_ZX_nothing_path)
uqc_1qubit_ZY_continuous = Analysis(uqc_1qubit_ZY_continuous_path)
uqc_1qubit_ZY_nothing = Analysis(uqc_1qubit_ZY_nothing_path)
```

```python
uqc_1qubit_XY_continuous_return = uqc_1qubit_XY_continuous.get_rewards()
uqc_1qubit_XY_nothing_return = uqc_1qubit_XY_nothing.get_rewards()
uqc_1qubit_ZX_continuous_return = uqc_1qubit_ZX_continuous.get_rewards()
uqc_1qubit_ZX_nothing_return = uqc_1qubit_ZX_nothing.get_rewards()
uqc_1qubit_ZY_continuous_return = uqc_1qubit_ZY_continuous.get_rewards()
uqc_1qubit_ZY_nothing_return = uqc_1qubit_ZY_nothing.get_rewards()

uqc_1qubit_XY_continuous_return_mean = np.mean(uqc_1qubit_XY_continuous_return, axis=0)
uqc_1qubit_XY_nothing_return_mean = np.mean(uqc_1qubit_XY_nothing_return, axis=0)
uqc_1qubit_ZX_continuous_return_mean = np.mean(uqc_1qubit_ZX_continuous_return, axis=0)
uqc_1qubit_ZX_nothing_return_mean = np.mean(uqc_1qubit_ZX_nothing_return, axis=0)
uqc_1qubit_ZY_continuous_return_mean = np.mean(uqc_1qubit_ZY_continuous_return, axis=0)
uqc_1qubit_ZY_nothing_return_mean = np.mean(uqc_1qubit_ZY_nothing_return, axis=0)

uqc_1qubit_XY_continuous_return_std = np.std(uqc_1qubit_XY_continuous_return, axis=0)
uqc_1qubit_XY_nothing_return_std = np.std(uqc_1qubit_XY_nothing_return, axis=0)
uqc_1qubit_ZX_continuous_return_std = np.std(uqc_1qubit_ZX_continuous_return, axis=0)
uqc_1qubit_ZX_nothing_return_std = np.std(uqc_1qubit_ZX_nothing_return, axis=0)
uqc_1qubit_ZY_continuous_return_std = np.std(uqc_1qubit_ZY_continuous_return, axis=0)
uqc_1qubit_ZY_nothing_return_std = np.std(uqc_1qubit_ZY_nothing_return, axis=0)
```

```python
fig, axs = plt.subplots(1, 3, figsize=(15, 5), tight_layout = True)

axs[0].plot(uqc_1qubit_XY_continuous_return_mean, color = "red", label="XY continuous")
axs[0].plot(uqc_1qubit_XY_nothing_return_mean, color = "blue", label="XY nothing")
axs[0].plot(uqc_1qubit_ZX_continuous_return_mean, color = "green", label="ZX continuous")
axs[0].plot(uqc_1qubit_ZX_nothing_return_mean, color = "orange", label="ZX nothing")
axs[0].plot(uqc_1qubit_ZY_continuous_return_mean, color = "purple", label="ZY continuous")
axs[0].plot(uqc_1qubit_ZY_nothing_return_mean, color = "brown", label="ZY nothing")
axs[0].set_xlabel("Episode")
axs[0].set_ylabel("Mean Return")
axs[0].set_title("All models")
axs[0].legend()

axs[1].plot(uqc_1qubit_XY_continuous_return_mean, color = "red", label="XY")
axs[1].fill_between(np.arange(len(uqc_1qubit_XY_continuous_return_mean)), uqc_1qubit_XY_continuous_return_mean - uqc_1qubit_XY_continuous_return_std, np.clip(uqc_1qubit_XY_continuous_return_mean + uqc_1qubit_XY_continuous_return_std, a_min = None, a_max = 200), alpha=0.15, color = "red")
axs[1].plot(uqc_1qubit_ZX_continuous_return_mean, color = "blue", label="ZX")
axs[1].fill_between(np.arange(len(uqc_1qubit_ZX_continuous_return_mean)), uqc_1qubit_ZX_continuous_return_mean - uqc_1qubit_ZX_continuous_return_std, np.clip(uqc_1qubit_ZX_continuous_return_mean + uqc_1qubit_ZX_continuous_return_std, a_min = None, a_max = 200), alpha=0.15, color = "blue")
axs[1].plot(uqc_1qubit_ZY_continuous_return_mean, color = "green", label="ZY")
axs[1].fill_between(np.arange(len(uqc_1qubit_ZY_continuous_return_mean)), uqc_1qubit_ZY_continuous_return_mean - uqc_1qubit_ZY_continuous_return_std, np.clip(uqc_1qubit_ZY_continuous_return_mean + uqc_1qubit_ZY_continuous_return_std, a_min = None, a_max = 200), alpha=0.15, color = "green")
axs[1].set_xlabel("Episode")
axs[1].set_ylabel("Mean Return")
axs[1].set_title("Continuous")
axs[1].legend()

axs[2].plot(uqc_1qubit_XY_nothing_return_mean, color = "red", label="XY")
axs[2].plot(uqc_1qubit_ZX_nothing_return_mean, color = "blue", label="ZX")
axs[2].plot(uqc_1qubit_ZY_nothing_return_mean, color = "green", label="ZY")
axs[2].fill_between(np.arange(len(uqc_1qubit_XY_nothing_return_mean)), uqc_1qubit_XY_nothing_return_mean - uqc_1qubit_XY_nothing_return_std, np.clip(uqc_1qubit_XY_nothing_return_mean + uqc_1qubit_XY_nothing_return_std, a_min = None, a_max = 200), alpha=0.15, color = "red")
axs[2].fill_between(np.arange(len(uqc_1qubit_ZX_nothing_return_mean)), uqc_1qubit_ZX_nothing_return_mean - uqc_1qubit_ZX_nothing_return_std, np.clip(uqc_1qubit_ZX_nothing_return_mean + uqc_1qubit_ZX_nothing_return_std, a_min = None, a_max = 200), alpha=0.15, color = "blue")
axs[2].fill_between(np.arange(len(uqc_1qubit_ZY_nothing_return_mean)), uqc_1qubit_ZY_nothing_return_mean - uqc_1qubit_ZY_nothing_return_std, np.clip(uqc_1qubit_ZY_nothing_return_mean + uqc_1qubit_ZY_nothing_return_std, a_min = None, a_max = 200), alpha=0.15, color = "green")
axs[2].set_xlabel("Episode")
axs[2].set_ylabel("Mean Return")
axs[2].set_title("Nothing")
axs[2].legend()

plt.show()


```
