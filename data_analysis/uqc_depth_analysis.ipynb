```python
import random
import numpy as np
import matplotlib.pyplot as plt
import pandas as pd
from analysis_functions import Analysis
```

```python
uqc_path = "../../results/uqc_thesis_testing/models_testing/universal_quantum_classifier"
multiqubit_path = "../../results/uqc_thesis_testing/models_testing/multiqubit"
fullencoding_path = "../../results/uqc_thesis_testing/models_testing/full_encoding"
full_encoding_tiledinputs_path = "../../results/uqc_thesis_testing/models_testing/full_encoding_tiledinputs"

uqc_analysis = Analysis(uqc_path)
multiqubit_analysis = Analysis(multiqubit_path)
fullencoding_analysis = Analysis(fullencoding_path)
full_encoding_tiledinputs_analysis = Analysis(full_encoding_tiledinputs_path)
```

```python
uqc_rewards = uqc_analysis.get_rewards()
multiqubit_rewards = multiqubit_analysis.get_rewards()
fullencoding_rewards = fullencoding_analysis.get_rewards()
full_encoding_tiledinputs_rewards = full_encoding_tiledinputs_analysis.get_rewards()

uqc_mean_rewards = np.mean(uqc_rewards, axis=0)
multiqubit_mean_rewards = np.mean(multiqubit_rewards, axis=0)
fullencoding_mean_rewards = np.mean(fullencoding_rewards, axis=0)
full_encoding_tiledinputs_mean_rewards = np.mean(full_encoding_tiledinputs_rewards, axis=0)

uqc_std_rewards = np.std(uqc_rewards, axis=0)
multiqubit_std_rewards = np.std(multiqubit_rewards, axis=0)
fullencoding_std_rewards = np.std(fullencoding_rewards, axis=0)
full_encoding_tiledinputs_std_rewards = np.std(full_encoding_tiledinputs_rewards, axis=0)
```

```python
plt.plot(uqc_mean_rewards, label="UQC")
plt.plot(multiqubit_mean_rewards, label="Multiqubit")
plt.plot(fullencoding_mean_rewards, label="Full encoding")
plt.plot(full_encoding_tiledinputs_mean_rewards, label="Full encoding tiled inputs")
plt.fill_between(range(len(uqc_mean_rewards)), uqc_mean_rewards - uqc_std_rewards, uqc_mean_rewards + uqc_std_rewards, alpha=0.2)
plt.fill_between(range(len(multiqubit_mean_rewards)), multiqubit_mean_rewards - multiqubit_std_rewards, multiqubit_mean_rewards + multiqubit_std_rewards, alpha=0.2)
plt.fill_between(range(len(fullencoding_mean_rewards)), fullencoding_mean_rewards - fullencoding_std_rewards, fullencoding_mean_rewards + fullencoding_std_rewards, alpha=0.2)
plt.fill_between(range(len(full_encoding_tiledinputs_mean_rewards)), full_encoding_tiledinputs_mean_rewards - full_encoding_tiledinputs_std_rewards, full_encoding_tiledinputs_mean_rewards + full_encoding_tiledinputs_std_rewards, alpha=0.2)
plt.legend()
plt.xlabel("Episode")
plt.ylabel("Reward")
plt.show()
```
