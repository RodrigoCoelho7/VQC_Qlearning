```python
import numpy as np
import matplotlib.pyplot as plt
import pandas as pd
from analysis_functions import Analysis
```

```python
w_mean0_std1_b0_path = "../../results/uqc_initialization/w_mean0_std1_b0"
w_means0_std001_path = "../../results/uqc_initialization/w_mean0_std0,01_b0"
w_means0_std001_bsame_path = "../../results/uqc_initialization/w_mean0_std0,01_bsame"
```

```python
w_mean0_std1_b0 = Analysis(w_mean0_std1_b0_path)
w_means0_std001 = Analysis(w_means0_std001_path)
w_means0_std001_bsame = Analysis(w_means0_std001_bsame_path)
```

```python
w_mean0_std1_b0_return = w_mean0_std1_b0.get_rewards()
w_mean0_std1_b0_mean_return = np.mean(w_mean0_std1_b0_return, axis=0)
w_mean0_std1_b0_std_return = np.std(w_mean0_std1_b0_return, axis=0)

w_means0_std001_return = w_means0_std001.get_rewards()
w_means0_std001_mean_return = np.mean(w_means0_std001_return, axis=0)
w_means0_std001_std_return = np.std(w_means0_std001_return, axis=0)

w_means0_std001_bsame_return = w_means0_std001_bsame.get_rewards()
w_means0_std001_bsame_mean_return = np.mean(w_means0_std001_bsame_return, axis=0)
w_means0_std001_bsame_std_return = np.std(w_means0_std001_bsame_return, axis=0)
```

```python
plt.plot(w_mean0_std1_b0_mean_return)
plt.fill_between(range(len(w_mean0_std1_b0_mean_return)), w_mean0_std1_b0_mean_return - w_mean0_std1_b0_std_return, w_mean0_std1_b0_mean_return + w_mean0_std1_b0_std_return, alpha=0.2)
plt.plot(w_means0_std001_mean_return)
plt.fill_between(range(len(w_means0_std001_mean_return)), w_means0_std001_mean_return - w_means0_std001_std_return, w_means0_std001_mean_return + w_means0_std001_std_return, alpha=0.2)
plt.plot(w_means0_std001_bsame_mean_return)
plt.fill_between(range(len(w_means0_std001_bsame_mean_return)), w_means0_std001_bsame_mean_return - w_means0_std001_bsame_std_return, w_means0_std001_bsame_mean_return + w_means0_std001_bsame_std_return, alpha=0.2)
```
